-- Consulta: Listar todos os alunos e as empresas onde estão a estagiar
-- Mostra o nome do aluno, o nome da empresa e as datas de início e fim do estágio
-- Ordenado pelo nome do aluno
-- =============================================

```python
query = """
SELECT 
    alunos.nome_aluno AS Aluno,
    empresas.nome_empresa AS Empresa,
    estagioo.data_inicio AS `Início do Estágio`,
    estagioo.data_fim AS `Fim do Estágio`
FROM estagioo
INNER JOIN alunos ON estagioo.id_aluno = alunos.id_aluno
INNER JOIN empresas ON estagioo.id_empresa = empresas.id_empresa
ORDER BY alunos.nome_aluno;
"""

df = pd.read_sql(query, conn)
df
```

### Consulta: Listar todos os estágios com detalhes completos

Esta query retorna informações completas sobre todos os estágios, incluindo:

- **Empresa**: Nome da empresa onde o estágio ocorre  
- **ID do Estágio**: Identificador único do estágio  
- **Aluno**: Nome do aluno que está a realizar o estágio  
- **Curso**: Curso do aluno  
- **Tutor**: Nome do tutor responsável  
- **Nota**: Nota final do estágio  
- **Início do Estágio**: Data de início do estágio  
- **Fim do Estágio**: Data de término do estágio  

Os resultados são **ordenados primeiro pelo nome da empresa e depois pelo ID do estágio**.


```python
query = """
SELECT 
    empresas.nome_empresa AS Empresa,
    estagioo.id_estagio AS `ID do Estágio`,
    alunos.nome_aluno AS Aluno,
    cursos.nome_curso AS Curso,
    tutores.nome_tutor AS Tutor,
    avaliacao.nota AS Nota,
    estagioo.data_inicio AS `Início do Estágio`,
    estagioo.data_fim AS `Fim do Estágio`
FROM estagioo
INNER JOIN alunos 
    ON estagioo.id_aluno = alunos.id_aluno
INNER JOIN cursos 
    ON alunos.id_curso = cursos.id_curso
INNER JOIN empresas 
    ON estagioo.id_empresa = empresas.id_empresa
INNER JOIN tutores 
    ON estagioo.id_tutor = tutores.id_tutor
INNER JOIN avaliacao 
    ON estagioo.id_aval = avaliacao.id_aval
ORDER BY empresas.nome_empresa, estagioo.id_estagio;
"""

df = pd.read_sql(query, conn)
df
```

### Consulta: Total de estágios por empresa

Esta query retorna o **número total de estágios realizados em cada empresa**:

- **Empresa**: Nome da empresa  
- **Total de Estágios**: Quantidade de estágios realizados nessa empresa  

Os resultados são **ordenados do maior para o menor número de estágios**.


```python
query = """
SELECT 
    empresas.nome_empresa AS Empresa,
    COUNT(estagioo.id_estagio) AS `Total de Estágios`
FROM estagioo
INNER JOIN empresas 
    ON estagioo.id_empresa = empresas.id_empresa
GROUP BY empresas.nome_empresa
ORDER BY `Total de Estágios` DESC;
"""

df = pd.read_sql(query, conn)
df
```

### Consulta: Média das notas dos estágios por empresa

Esta query retorna a **média das notas dos estágios** para cada empresa:

- **Empresa**: Nome da empresa  
- **Média das Notas**: Média das avaliações dos estágios, arredondada para 2 casas decimais  

Os resultados são **ordenados da maior para a menor média de notas**.


```python
query = """
SELECT 
    empresas.nome_empresa AS Empresa,
    ROUND(AVG(avaliacao.nota), 2) AS `Média das Notas`
FROM estagioo
INNER JOIN empresas 
    ON estagioo.id_empresa = empresas.id_empresa
INNER JOIN avaliacao 
    ON estagioo.id_aval = avaliacao.id_aval
GROUP BY empresas.nome_empresa
ORDER BY `Média das Notas` DESC;
"""

df = pd.read_sql(query, conn)
df
```

### Consulta: Média combinada entre nota do curso e nota do estágio

Esta query calcula a **média entre a nota do curso e a nota final do estágio** para cada aluno:

- **Aluno**: Nome do aluno  
- **Curso**: Nome do curso do aluno  
- **Nota do Curso**: Nota obtida pelo aluno no curso  
- **Nota da Avaliação**: Nota obtida no estágio  
- **Média Combinada**: Média entre a nota do curso e a nota do estágio, arredondada para 2 casas decimais  

Os resultados são **ordenados da maior para a menor média combinada**.


```python
query = """
SELECT 
    alunos.nome_aluno AS Aluno,
    cursos.nome_curso AS Curso,
    alunos.nota_curso AS `Nota do Curso`,
    avaliacao.nota AS `Nota da Avaliação`,
    ROUND((alunos.nota_curso + avaliacao.nota) / 2, 2) AS `Média Combinada`
FROM estagioo
INNER JOIN alunos 
    ON estagioo.id_aluno = alunos.id_aluno
INNER JOIN cursos 
    ON alunos.id_curso = cursos.id_curso
INNER JOIN avaliacao 
    ON estagioo.id_aval = avaliacao.id_aval
ORDER BY `Média Combinada` DESC;
"""

df = pd.read_sql(query, conn)
df
```

### Consulta: Total de alunos por tutor e empresa

Esta query retorna o **número de alunos que cada tutor acompanha em cada empresa**:

- **Empresa**: Nome da empresa  
- **Tutor**: Nome do tutor  
- **Total de Alunos**: Quantidade de alunos supervisionados pelo tutor na empresa  

Os resultados são **ordenados do maior para o menor número de alunos**.


```python
query = """
SELECT 
    empresas.nome_empresa AS Empresa,
    tutores.nome_tutor AS Tutor,
    COUNT(estagioo.id_estagio) AS `Total de Alunos`
FROM estagioo
INNER JOIN tutores 
    ON estagioo.id_tutor = tutores.id_tutor
INNER JOIN empresas 
    ON tutores.id_empresa = empresas.id_empresa
GROUP BY empresas.nome_empresa, tutores.nome_tutor
ORDER BY `Total de Alunos` DESC;
"""

df = pd.read_sql(query, conn)
df
```

### Consulta: Alunos com nota do estágio inferior a 12

Esta query retorna todos os alunos que **obtiveram nota inferior a 12 na avaliação do estágio**:

- **Aluno**: Nome do aluno  
- **Curso**: Curso do aluno  
- **Nota**: Nota obtida na avaliação do estágio  

Os resultados são **ordenados pelo nome do curso e depois pela nota em ordem crescente**.


```python
query = """
SELECT 
    alunos.nome_aluno AS Aluno,
    cursos.nome_curso AS Curso,
    avaliacao.nota AS Nota
FROM estagioo
INNER JOIN alunos 
    ON estagioo.id_aluno = alunos.id_aluno
INNER JOIN cursos 
    ON alunos.id_curso = cursos.id_curso
INNER JOIN avaliacao 
    ON estagioo.id_aval = avaliacao.id_aval
WHERE avaliacao.nota < 12
ORDER BY cursos.nome_curso, Nota ASC;
"""

df = pd.read_sql(query, conn)
df
```
